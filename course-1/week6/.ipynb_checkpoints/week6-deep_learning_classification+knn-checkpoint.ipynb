```python
import graphlab
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
image_train = graphlab.SFrame('image_train_data/')
```

```python
image_test = graphlab.SFrame('image_test_data/')
```

```python
image_train.head()
```

# Train a classifier on raw image pixels

```python
raw_pixel_model = graphlab.logistic_classifier.create(image_train, target = 'label', features=['image_array'])
```

```python
image_test[0:3]['image'].show()
```

```python
image_test[0:3].show()
```

```python
raw_pixel_model.predict(image_test[0:3])
```

```python
raw_pixel_model.evaluate(image_test)
```

```python
deep_feature_model = graphlab.logistic_classifier.create(image_train, target='label',features=['deep_features'],
                                                        max_iterations = 10)
```

```python
deep_feature_model.predict(image_test[0:3])
```

```python
deep_feature_model.evaluate(image_test)
```

# Train KNN for similar images

```python
knn_model = graphlab.nearest_neighbors.create(image_train, features=['deep_features'], label='id')
```

```python
car = image_train[8:9]
```

```python
knn_model.query(car)
```

```python
image_train.filter_by(knn_model.query(car)['reference_label'], 'id')['image'].show()
```

```python
def similar_images(i):
    image_train.filter_by(knn_model.query(image_train[i:i+1])['reference_label'], 'id')['image'].show()
```

```python
similar_images(8)
```
